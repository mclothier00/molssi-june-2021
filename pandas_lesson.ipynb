## Accessing Data

```python
# To slice a pandas dataframe using row and column numbers
# dataframe_name.iloc[row_slice, column_slice]

periodic_data.iloc[35]
```

```python
periodic_data.iloc[35,:3]
```

```python
periodic_data.iloc[35:45,:3]
```

```python
# Use the iloc function to:
# 1. Select the 5th row
# 2. Select row 30 to the end and the first three columns.
# 3. Select row 50 to the end and every other row and columns 2 through 4

periodic_data.iloc[5]
```

```python
periodic_data.iloc[30:,0:3]
```

```python
periodic_data.iloc[50::2,2:4]
# [starting row/column:endingrow/column:how many to skip]
```

```python
# will list column names
periodic_data.columns 
```

```python
# Third way to get information from dataframes - loc function
# loc functions lets us use column and row names

periodic_data.loc[35, "Symbol"]
```
