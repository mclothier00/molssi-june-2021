```python
import glob # glob is a python library like os
```

```python
file_location = os.path.join('data','outfiles','*.out') #* is a wild card
print(file_location)
```

```python
filenames = glob.glob(file_location)
print(filenames)
```

```python
for file in filenames:
    outfile = open(file,'r')
    data = outfile.readlines()
    outfile.close() # close file because will not be able to open a new file 
    for line in data:
        if 'Final Energy' in line:
            energy_line = line
            words = energy_line.split()
            energy = float(words[3])
            print(energy)
```

```python
for file in filenames:
    file_name = os.path.basename(file)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]
    
    outfile = open(file,'r')
    data = outfile.readlines()
    outfile.close() # close file because will not be able to open a new file 
    for line in data:
        if "Final Energy" in line:
            energy_line = line
            words = energy_line.split()
            energy = float(words[3])
            print(F'{molecule_name}: {energy}')
```

optional:
- can format how energy prints decimals:
    print(F'{molecule_name}: {energy:.4f}')
- can include a tab instead of a colon: 
    print(F'{molecule_name} \t {energy}')
- new line is \n

```python
datafile = open('energies.txt', 'w+') # w+ creates file if it does not exist
print('Hello')
for file in filenames:
    file_name = os.path.basename(file)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]
    
    outfile = open(file,'r')
    data = outfile.readlines()
    outfile.close() # close file because will not be able to open a new file 
    for line in data:
        if "Final Energy" in line:
            energy_lines = line
            words = energy_lines.split()
            energy = float(words[3])
            print(energy)
            datafile.write(F'{molecule_name} \t {energy:.4f} \n') # does not actually write to file... until close datafile
datafile.close()
```
