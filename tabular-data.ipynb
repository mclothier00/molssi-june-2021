# Working with Tabular Data
Tabular data lesson; day two of MolSSI workshop.

```python
import os
import numpy
```

```python
distance_file = os.path.join('data','distance_data_headers.csv')
print(distance_file)
```

```python
# use numpy function genfromtxt to read in the data
help(numpy.genfromtxt)
```

```python
distances = numpy.genfromtxt(fname=distance_file, delimiter = ',', dtype='unicode') 
# store output from genfromtxt
# unicode good with numbers and letters
# computer still thinks numbers are strings so two different variables
```

```python
print(distances)
```

```python
headers = distances[0]
print(headers)
```

```python
# Take slice of the data that is just the numerical values. Call slice data.
data = distances[1:]
print(data)
```

```python
data = data.astype(numpy.float)
```

```python
print(data)
```

```python
# array_name[row,column]
print(data[2,1])
```

```python
#What will the output of this code? (ans: 8.95420, 2.0)
element1 = data[0,1]
element2 = data[1,0]
#my code:
print(element1)
print(element2)

#her code:
print(F'The value of element 1 is {element1}.')
print(F'The value of element 2 is {element2}.')
```

```python
# new_array = array_name[start:end,start:end]
small_data = data[0:10,0:3]
print(small_data)
```

```python
# What are these slices? Write print statement to check.
array1 = small_data[5,:]
array2 = small_data[:,1:]
# my code
print(array1)
print(array2)

# her code
print(F'The array1 is {array1}.')
print(array2)
```

```python
print(data)
```

```python
thr4_atp = data[:,1]
average = numpy.mean(thr4_atp)
print(average)
```

```python
# range for loop: range(start,end)
```

```python
# calculate the number of columns
num_columns = len(headers) #or
num_columns = len(data[0,:])
print(num_columns)
```

```python
for i in range(1,num_columns):
    column = data[:,i]
    avg_col = numpy.mean(column)
    print(F'{headers[i]}: {avg_col:.3f}')
```

## Plotting
Creating plots using matplotlib.

```python
import matplotlib.pyplot as plt # plt is a nickname for library
```

```python
plt.figure()
plt.plot(data[:,1]) # do not need x value in this circumstance
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,1])
```

```python
# Pull out name of first column of data and save with variable name sample.
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,1], label=sample)
plt.legend()

# to save the figure:
plt.savefig(F'{sample}.png')
```

```python
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,1], label=sample)
plt.legend()

# to save a high quality figure (for publication or presentation):
plt.savefig(F'{sample}_high_quality.png',dpi=300) # 300 usually asked for presentations
```

```python
# plot two sets of data on the same graph
sample = headers[1]
sample2 = headers[2]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,1], label=sample)
plt.plot(data[:,2], label=sample2)
plt.legend()
```

```python
plt.figure() # not in for loop bc only need to do once

for i in range(1,num_columns):
    sample = headers[i]
    
    plt.plot(data[:,i], label=sample)
    
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()
```

```python
# modify for loop to create plot for each column

for i in range(1,num_columns):
    plt.figure() 
    sample = headers[i]
    plt.plot(data[:,i], label=sample)
    
    plt.ylim(4,20) # make all y-axis range from 4 to 20
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    
    plt.savefig(f'{sample}.png')
```
