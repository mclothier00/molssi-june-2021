```python
md_file = os.path.join('03_Prod.mdout')
print(md_file)
```

```python
outfile = open(md_file, 'r')
data = outfile.readlines()
outfile.close()

for line in data:
    if 'Etot' in line:
        energy_line = line
        print(energy_line)
```

```python
words = energy_line.split()
print(words)
```

```python
outfile = open(md_file, 'r')
data = outfile.readlines()
outfile.close()

datafile = open('md_energies.txt','w+')
for line in data:
    if 'Etot' in line:
        energy_line = line
        words = energy_line.split()
        energy = float(words[2])
        datafile.write(F'Etot = \t {energy} \n')
datafile.close()
```
